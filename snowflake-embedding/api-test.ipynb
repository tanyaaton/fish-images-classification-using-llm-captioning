```python

import requests
BASE_URL = "http://localhost:8080/extract_text"

def scoring_embedder(sentence):
    payload = {
        "sentence": [sentence]
    }
    response = requests.post(BASE_URL, json=payload)
    response_json = response.json()
    return response_json["predictions"][0]["values"][0][1]

scoring_embedder("Hello world!")
```
